```python
class Graphs:
    #deep Narayan
    #chaurasiya
    def __init__(self, size):
        self.size = size
        self.adj_list = [[] for _ in range(self.size)]

    def create_node(self, data):
        if data not in self.adj_list:
            self.adj_list[data] = []

    def push(self, node1, node2):
        if node1 not in self.adj_list:
            self.create_node(node1)
        if node2 not in self.adj_list:
            self.create_node(node2)

        self.adj_list[node1].append(node2)
        self.adj_list[node2].append(node1)

    def display(self):
        for i in self.adj_list:
            print(i)
g = Graphs(3)
g.push(10, 20)
g.push(20, 10)
g.push(30, 20)
g.display()
```
